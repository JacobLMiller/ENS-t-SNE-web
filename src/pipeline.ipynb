```python
import numpy as np 
import pandas as pd 
import json 
```

```python
from application.library.enstsne import load_penguins, ENSTSNE, load_clusters
```

```python
dists, labels, X = load_clusters()

enstsne = ENSTSNE(dists,30,labels)
enstsne.gd(5000)
```

```python
obj = {
    "data": list(),
    "projections": list()
}
```

```python
X = enstsne.get_embedding()
for i,el in enumerate(X):
    obj["data"].append({
        "x3": el[0],
        "y3": el[1],
        "z3": el[2],
        "class": [int(labels[0][i]), int(labels[1][i])]
    })
type(obj["data"][0]["class"][0])
```

```python
obj["projections"] = [a.tolist() for a in enstsne.P]
```

```python
with open("application/static/data/penguins.json", "w") as fdata:
    json.dump(obj, fdata, indent=4)
```
